```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn import metrics
from sklearn.decomposition import PCA
```

```python
df = np.array(pd.read_csv('HTRU/HTRU_2.csv',sep=',',header=0))
y = df[:,-1]
y = y.reshape((y.shape[0],1))
x = df[:,:-1]

gt_0 = np.where(y==0)[0]
gt_1 = np.where(y==1)[0]

x = StandardScaler().fit_transform(x)
```

```python
affinities = ["euclidean", "l1", "l2", "manhattan", "cosine"]
linkages = ["ward", "complete", "average", "single"]
```

```python
for link in linkages:
    for affs in affinities:
        if link == 'ward' and affs != 'euclidean':
            continue
        print("\nLinkage,Affinity = ",link,",",affs,"\n")

        components= [2,3,4,5,6,7]
        for i in components:
            x_PCA = PCA(n_components=i).fit_transform(x)
            clustering = AgglomerativeClustering(n_clusters = 2, affinity = affs, linkage = link).fit(x_PCA)
            y_pred = clustering.labels_

            pos_0 = np.where(y_pred==0)[0]
            pos_1 = np.where(y_pred==1)[0]

            if len(pos_0)<len(pos_1):
                temp = pos_0
                pos_0 = pos_1
                pos_1 = temp
                y_pred[pos_0] = 0
                y_pred[pos_1] = 1

            count_0 = 0
            count_1 = 0

            for j in pos_0:
                if y_pred[j] == y[j]:
                    count_0 +=1

            for j in pos_1:
                if y_pred[j] == y[j]:
                    count_1 +=1


            print(i,round(count_0*100/len(gt_0),2),round(count_1*100/len(gt_1),2),round((count_1+count_0)*100/(len(gt_1)+len(gt_0)),2))
        
        clustering = AgglomerativeClustering(n_clusters = 2, affinity = affs, linkage = link).fit(x_PCA)
        y_pred = clustering.labels_

        pos_0 = np.where(y_pred==0)[0]
        pos_1 = np.where(y_pred==1)[0]

        if len(pos_0)<len(pos_1):
            temp = pos_0
            pos_0 = pos_1
            pos_1 = temp
            y_pred[pos_0] = 0
            y_pred[pos_1] = 1

        count_0 = 0
        count_1 = 0

        for j in pos_0:
            if y_pred[j] == y[j]:
                count_0 +=1

        for j in pos_1:
            if y_pred[j] == y[j]:
                count_1 +=1


        print("Without PCA",round(count_0*100/len(gt_0),2),round(count_1*100/len(gt_1),2),round((count_1+count_0)*100/(len(gt_1)+len(gt_0)),2))
```
